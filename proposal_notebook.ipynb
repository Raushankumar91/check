```python
def find_n(p,q):
    m=len(p);
    n=len(q);
    u=0;
    for i in range(m):
        for j in range(n):
            if(i+j>=m):
                break;
            if(p[i+j]==q[j]):
                continue;
            break;
        if(j==m):
            return 1;
    return 0;
```

```python
def find1(u,df):
    for i in df.index:
        st=df.at[i,'Heading'];
        #print(st,u,type(u))
        if(type(u)==float):
           continue;
        if(st.lower()==u.lower()):
            print('djk')
            return[df.at[i,'Data type'],df.at[i,'Lengths']]
        else:
            p=st.lower();
            q=u.lower();
            if(p[0]==q[0]):
                vaq=find_n(p,q);
                if(vaq==1):
                    print('djk')
                    return[df.at[i,'Data type'],df.at[i,'Lengths']]
                    
    return['-','-']
```
